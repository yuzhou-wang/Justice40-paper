```python
def calculate_margin_value(data):
    data["margin_p"] = data["conc_p"]/data["emis"]
    data["margin_j40_tract"] = data["conc_j40_v1_tract"]/data["emis"]
    data["margin_outside_tract"] = data["conc_outside_v1_tract"]/data["emis"]
    data["margin_w"] = data["conc_w"]/data["emis"]
    data["margin_b"] = data["conc_b"]/data["emis"]
    data["margin_a"] = data["conc_a"]/data["emis"]
    data["margin_hw"] = data["conc_hw"]/data["emis"]
    data["margin_h"] = data["conc_h"]/data["emis"]
    data["margin_m"] = data["conc_m"]/data["emis"]
    data["margin_o"] = data["conc_o"]/data["emis"]
    data["margin_n"] = data["conc_n"]/data["emis"]
    data["margin_lowinc"] = data["conc_lowinc"]/data["emis"]
    data["margin_highinc"] = data["conc_highinc"]/data["emis"]

    data["margin_p_urban"] = data["conc_p_urban"]/data["emis"]
    data["margin_j40_tract_urban"] = data["conc_j40_tract_urban"]/data["emis"]
    data["margin_outside_tract_urban"] = data["conc_outside_tract_urban"]/data["emis"]
    data["margin_w_urban"] = data["conc_w_urban"]/data["emis"]
    data["margin_b_urban"] = data["conc_b_urban"]/data["emis"]
    data["margin_a_urban"] = data["conc_a_urban"]/data["emis"]
    data["margin_hw_urban"] = data["conc_hw_urban"]/data["emis"]
    data["margin_h_urban"] = data["conc_h_urban"]/data["emis"]
    data["margin_m_urban"] = data["conc_m_urban"]/data["emis"]
    data["margin_o_urban"] = data["conc_o_urban"]/data["emis"]
    data["margin_n_urban"] = data["conc_n_urban"]/data["emis"]
    data["margin_lowinc_urban"] = data["conc_lowinc_urban"]/data["emis"]
    data["margin_highinc_urban"] = data["conc_highinc_urban"]/data["emis"]

    data["margin_p_rural"] = data["conc_p_rural"]/data["emis"]
    data["margin_j40_tract_rural"] = data["conc_j40_tract_rural"]/data["emis"]
    data["margin_outside_tract_rural"] = data["conc_outside_tract_rural"]/data["emis"]
    data["margin_w_rural"] = data["conc_w_rural"]/data["emis"]
    data["margin_b_rural"] = data["conc_b_rural"]/data["emis"]
    data["margin_a_rural"] = data["conc_a_rural"]/data["emis"]
    data["margin_hw_rural"] = data["conc_hw_rural"]/data["emis"]
    data["margin_h_rural"] = data["conc_h_rural"]/data["emis"]
    data["margin_m_rural"] = data["conc_m_rural"]/data["emis"]
    data["margin_o_rural"] = data["conc_o_rural"]/data["emis"]
    data["margin_n_rural"] = data["conc_n_rural"]/data["emis"]
    data["margin_lowinc_rural"] = data["conc_lowinc_rural"]/data["emis"]
    data["margin_highinc_rural"] = data["conc_highinc_rural"]/data["emis"]
    data["sector"]=np.where(data["sector"]=='Non-coal Elec', 'Non-coal Elec.',data["sector"])
    return(data)

```

```python
def sector_pollutant_reduction(data,percent_base):
    data_yearly = data.copy()
    year_list = [0]
    reduce_w = [data.conc_w.sum()]
    reduce_b = [data.conc_b.sum()]
    reduce_a = [data.conc_a.sum()]
    reduce_hw = [data.conc_hw.sum()]
    reduce_h = [data.conc_h.sum()]
    reduce_m = [data.conc_m.sum()]
    reduce_o = [data.conc_o.sum()]
    reduce_n = [data.conc_n.sum()]
    reduce_p = [data.conc_p.sum()]
    reduce_j40 = [data.conc_j40_v1_tract.sum()]
    reduce_outside = [data.conc_outside_v1_tract.sum()]
    reduce_lowinc = [data.conc_lowinc.sum()]
    reduce_highinc = [data.conc_highinc.sum()]

    reduce_w_urban = [data.conc_w_urban.sum()]
    reduce_b_urban = [data.conc_b_urban.sum()]
    reduce_a_urban = [data.conc_a_urban.sum()]
    reduce_hw_urban = [data.conc_hw_urban.sum()]
    reduce_h_urban = [data.conc_h_urban.sum()]
    reduce_m_urban = [data.conc_m_urban.sum()]
    reduce_o_urban = [data.conc_o_urban.sum()]
    reduce_n_urban = [data.conc_n_urban.sum()]
    reduce_p_urban = [data.conc_p_urban.sum()]
    reduce_j40_urban = [data.conc_j40_tract_urban.sum()]
    reduce_outside_urban = [data.conc_outside_tract_urban.sum()]
    reduce_lowinc_urban = [data.conc_lowinc_urban.sum()]
    reduce_highinc_urban = [data.conc_highinc_urban.sum()]

    reduce_w_rural = [data.conc_w_rural.sum()]
    reduce_b_rural = [data.conc_b_rural.sum()]
    reduce_a_rural = [data.conc_a_rural.sum()]
    reduce_hw_rural = [data.conc_hw_rural.sum()]
    reduce_h_rural = [data.conc_h_rural.sum()]
    reduce_m_rural = [data.conc_m_rural.sum()]
    reduce_o_rural = [data.conc_o_rural.sum()]
    reduce_n_rural = [data.conc_n_rural.sum()]
    reduce_p_rural = [data.conc_p_rural.sum()]
    reduce_j40_rural = [data.conc_j40_tract_rural.sum()]
    reduce_outside_rural = [data.conc_outside_tract_rural.sum()]
    reduce_lowinc_rural = [data.conc_lowinc_rural.sum()]
    reduce_highinc_rural = [data.conc_highinc_rural.sum()]

    emis_reduce = [0]

    conc_w = data.conc_w.sum()
    conc_b = data.conc_b.sum()
    conc_a = data.conc_a.sum()
    conc_hw = data.conc_hw.sum()
    conc_h = data.conc_h.sum()
    conc_m = data.conc_m.sum()
    conc_n = data.conc_n.sum()
    conc_o = data.conc_o.sum()
    conc_p = data.conc_p.sum()
    conc_j40 = data.conc_j40_v1_tract.sum()
    conc_outside = data.conc_outside_v1_tract.sum()
    conc_lowinc = data.conc_lowinc.sum()
    conc_highinc = data.conc_highinc.sum()

    conc_w_urban = data.conc_w_urban.sum()
    conc_b_urban = data.conc_b_urban.sum()
    conc_a_urban = data.conc_a_urban.sum()
    conc_hw_urban = data.conc_hw_urban.sum()
    conc_h_urban = data.conc_h_urban.sum()
    conc_m_urban = data.conc_m_urban.sum()
    conc_n_urban = data.conc_n_urban.sum()
    conc_o_urban = data.conc_o_urban.sum()
    conc_p_urban = data.conc_p_urban.sum()
    conc_j40_urban = data.conc_j40_tract_urban.sum()
    conc_outside_urban = data.conc_outside_tract_urban.sum()
    conc_lowinc_urban = data.conc_lowinc_urban.sum()
    conc_highinc_urban = data.conc_highinc_urban.sum()

    conc_w_rural = data.conc_w_rural.sum()
    conc_b_rural = data.conc_b_rural.sum()
    conc_a_rural = data.conc_a_rural.sum()
    conc_hw_rural = data.conc_hw_rural.sum()
    conc_h_rural = data.conc_h_rural.sum()
    conc_m_rural = data.conc_m_rural.sum()
    conc_n_rural = data.conc_n_rural.sum()
    conc_o_rural = data.conc_o_rural.sum()
    conc_p_rural = data.conc_p_rural.sum()
    conc_j40_rural = data.conc_j40_tract_rural.sum()
    conc_outside_rural = data.conc_outside_tract_rural.sum()
    conc_lowinc_rural = data.conc_lowinc_rural.sum()
    conc_highinc_rural = data.conc_highinc_rural.sum()

    emis_reduce_value = 0
    emission_sector = [data_yearly["emis"].to_list()]


    emis_reduce_total = 1
    year = 1
    #data_yearly["emis_reduce"] = data_yearly["emis"]*data_yearly["percentage"]/100*times
    data_yearly["percent_times"] = data_yearly[percent_base]*1e6
    data_yearly["emis_reduce"] = data_yearly[["emis","percent_times"]].min(axis = 1)
    #yearly_emission_value = data_yearly["emis"].sum()*0.0344*times
    
    #yearly_emission_value = yearly_reduction*1000
    data_yearly["emis_reduce_step"] = data_yearly["emis_reduce"].values
    while year < 21:
        emis_reduce_value += data_yearly["emis_reduce_step"].sum()
        conc_w -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_w"])
        conc_b -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_b"])
        conc_a -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_a"])
        conc_hw -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_hw"])
        conc_h -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_h"])
        conc_m -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_m"])
        conc_n -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_n"])
        conc_o -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_o"])
        conc_p -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_p"])
        conc_j40 -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_j40_tract"])
        conc_outside -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_outside_tract"])
        conc_lowinc -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_lowinc"])
        conc_highinc -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_highinc"])

        conc_w_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_w_urban"])
        conc_b_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_b_urban"])
        conc_a_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_a_urban"])
        conc_hw_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_hw_urban"])
        conc_h_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_h_urban"])
        conc_m_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_m_urban"])
        conc_n_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_n_urban"])
        conc_o_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_o_urban"])
        conc_p_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_p_urban"])
        conc_j40_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_j40_tract_urban"])
        conc_outside_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_outside_tract_urban"])
        conc_lowinc_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_lowinc_urban"])
        conc_highinc_urban -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_highinc_urban"])

        conc_w_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_w_rural"])
        conc_b_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_b_rural"])
        conc_a_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_a_rural"])
        conc_hw_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_hw_rural"])
        conc_h_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_h_rural"])
        conc_m_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_m_rural"])
        conc_n_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_n_rural"])
        conc_o_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_o_rural"])
        conc_p_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_p_rural"])
        conc_j40_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_j40_tract_rural"])
        conc_outside_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_outside_tract_rural"])
        conc_lowinc_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_lowinc_rural"])
        conc_highinc_rural -= sum(data_yearly["emis_reduce_step"]*data_yearly["margin_highinc_rural"])

        # update emission value
        data_yearly["emis"] = data_yearly["emis"] -data_yearly["emis_reduce_step"]
        data_yearly["emis_reduce_step"] = data_yearly[['emis_reduce','emis']].min(axis=1)
        year_list.append(year)
        reduce_w.append(conc_w)
        reduce_b.append(conc_b)
        reduce_a.append(conc_a)
        reduce_hw.append(conc_hw)
        reduce_h.append(conc_h)
        reduce_m.append(conc_m)
        reduce_n.append(conc_n)
        reduce_o.append(conc_o)
        reduce_p.append(conc_p)
        reduce_j40.append(conc_j40)
        reduce_outside.append(conc_outside)
        reduce_lowinc.append(conc_lowinc)
        reduce_highinc.append(conc_highinc)

        reduce_w_urban.append(conc_w_urban)
        reduce_b_urban.append(conc_b_urban)
        reduce_a_urban.append(conc_a_urban)
        reduce_hw_urban.append(conc_hw_urban)
        reduce_h_urban.append(conc_h_urban)
        reduce_m_urban.append(conc_m_urban)
        reduce_n_urban.append(conc_n_urban)
        reduce_o_urban.append(conc_o_urban)
        reduce_p_urban.append(conc_p_urban)
        reduce_j40_urban.append(conc_j40_urban)
        reduce_outside_urban.append(conc_outside_urban)
        reduce_lowinc_urban.append(conc_lowinc_urban)
        reduce_highinc_urban.append(conc_highinc_urban)

        reduce_w_rural.append(conc_w_rural)
        reduce_b_rural.append(conc_b_rural)
        reduce_a_rural.append(conc_a_rural)
        reduce_hw_rural.append(conc_hw_rural)
        reduce_h_rural.append(conc_h_rural)
        reduce_m_rural.append(conc_m_rural)
        reduce_n_rural.append(conc_n_rural)
        reduce_o_rural.append(conc_o_rural)
        reduce_p_rural.append(conc_p_rural)
        reduce_j40_rural.append(conc_j40_rural)
        reduce_outside_rural.append(conc_outside_rural)
        reduce_lowinc_rural.append(conc_lowinc_rural)
        reduce_highinc_rural.append(conc_highinc_rural)

        emis_reduce.append(emis_reduce_value)
        year += 1
        emis_reduce_total = data_yearly["emis_reduce_step"].sum()
        emission_sector.append(data_yearly["emis"].to_list())
    #print(reduce_p)

    percentage_reduce_conc_year = pd.DataFrame({"year":year_list,"reduce_w":reduce_w,"reduce_b":reduce_b,"reduce_hw":reduce_hw
                                                            ,"reduce_h":reduce_h,"reduce_m":reduce_m,"reduce_n":reduce_n
                                                             ,"reduce_a":reduce_a,"reduce_o":reduce_o,"reduce_p":reduce_p,"reduce_j40":reduce_j40
                                               ,"reduce_outside":reduce_outside,"reduce_lowinc":reduce_lowinc,"reduce_highinc":reduce_highinc
                                                             ,"reduce_w_urban":reduce_w_urban,"reduce_b_urban":reduce_b_urban,"reduce_hw_urban":reduce_hw_urban,
                                                             "reduce_h_urban":reduce_h_urban,"reduce_m_urban":reduce_m_urban,"reduce_n_urban":reduce_n_urban,
                                                             "reduce_a_urban":reduce_a_urban,"reduce_o_urban":reduce_o_urban,"reduce_p_urban":reduce_p_urban,
                                                             "reduce_j40_urban":reduce_j40_urban,"reduce_outside_urban":reduce_outside_urban
                                                             ,"reduce_lowinc_urban":reduce_lowinc_urban,"reduce_highinc_urban":reduce_highinc_urban
                                                             ,"reduce_w_rural":reduce_w_rural,"reduce_b_rural":reduce_b_rural,"reduce_hw_rural":reduce_hw_rural,
                                                             "reduce_h_rural":reduce_h_rural,"reduce_m_rural":reduce_m_rural,"reduce_n_rural":reduce_n_rural,
                                                             "reduce_a_rural":reduce_a_rural,"reduce_o_rural":reduce_o_rural,"reduce_p_rural":reduce_p_rural,
                                                             "reduce_j40_rural":reduce_j40_rural,"reduce_outside_rural":reduce_outside_rural,
                                                        "reduce_lowinc_rural":reduce_lowinc_rural,"reduce_highinc_rural":reduce_highinc_rural, "emis":emis_reduce})
    percentage_reduce_conc_year["disparity_j40"] = percentage_reduce_conc_year["reduce_j40"]-percentage_reduce_conc_year["reduce_p"]
    
    percentage_reduce_conc_year["disparity"] = percentage_reduce_conc_year[["reduce_w","reduce_h","reduce_b","reduce_a","reduce_o","reduce_m"]].max(axis=1)-percentage_reduce_conc_year["reduce_p"]
    percentage_reduce_conc_year["disparity_j40_urban"] = percentage_reduce_conc_year["reduce_j40_urban"]-percentage_reduce_conc_year["reduce_p_urban"]
    percentage_reduce_conc_year["disparity_urban"] = percentage_reduce_conc_year[["reduce_w_urban","reduce_h_urban","reduce_b_urban","reduce_a_urban","reduce_o_urban","reduce_m_urban"]].max(axis=1)-percentage_reduce_conc_year["reduce_p_urban"]
    percentage_reduce_conc_year["disparity_j40_rural"] = percentage_reduce_conc_year["reduce_j40_rural"]-percentage_reduce_conc_year["reduce_p_rural"]
    percentage_reduce_conc_year["disparity_rural"] = percentage_reduce_conc_year[["reduce_w_rural","reduce_h_rural","reduce_b_rural","reduce_a_rural","reduce_o_rural","reduce_m_rural"]].max(axis=1)-percentage_reduce_conc_year["reduce_p_rural"]    
    percentage_reduce_conc_year["disparity_lowinc"] = percentage_reduce_conc_year["reduce_lowinc"]-percentage_reduce_conc_year["reduce_p"]
    percentage_reduce_conc_year["disparity_lowinc_urban"] = percentage_reduce_conc_year["reduce_lowinc_urban"]-percentage_reduce_conc_year["reduce_p_urban"]
    percentage_reduce_conc_year["disparity_lowinc_rural"] = percentage_reduce_conc_year["reduce_lowinc_rural"]-percentage_reduce_conc_year["reduce_p_rural"]
    
    percentage_reduce_conc_year["reduce_meg"] = percentage_reduce_conc_year[["reduce_w","reduce_h","reduce_b","reduce_a","reduce_o","reduce_m"]].max(axis=1)
    percentage_reduce_conc_year["reduce_meg_urban"] = percentage_reduce_conc_year[["reduce_w_urban","reduce_h_urban","reduce_b_urban","reduce_a_urban","reduce_o_urban","reduce_m_urban"]].max(axis=1)
    percentage_reduce_conc_year["reduce_meg_rural"] = percentage_reduce_conc_year[["reduce_w_rural","reduce_h_rural","reduce_b_rural","reduce_a_rural","reduce_o_rural","reduce_m_rural"]].max(axis=1)
    percentage_reduce_conc_year["reduce_poc"] = (percentage_reduce_conc_year["reduce_p"]*306675006-percentage_reduce_conc_year["reduce_w"]*196052889)/110622117
    percentage_reduce_conc_year["disparity_poc"] = percentage_reduce_conc_year["reduce_poc"]-percentage_reduce_conc_year["reduce_p"]

    emission_sector_df = pd.DataFrame(np.array(emission_sector).T.tolist())
    emission_sector_df["sector"] = data_yearly["sector"]
    emission_sector_df["pollutant"] = data_yearly["pollutant"]

    return(percentage_reduce_conc_year,emission_sector_df)

```
